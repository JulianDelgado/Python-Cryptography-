```python
# AES FUNCTIONS
def sub_bytes(s_matrix: pd.DataFrame):
    s_box = s_box_generate()
    s_prima_matrix_np = []
    s_sequence = matrix_to_sequence(s_matrix)
    s_array = re.findall('..', s_sequence)
    for byte in s_array:
        left_bits = byte[0]
        right_bits = byte[1]
        converted_value = s_box.loc[left_bits, right_bits]
        s_prima_matrix_np.append(converted_value)
    s_prima_matrix_np = np.array(s_prima_matrix_np)
    s_prima_matrix_np = np.reshape(s_prima_matrix_np, (4, 4))
    s_prima_matrix = pd.DataFrame(s_prima_matrix_np)
    return s_prima_matrix


def inv_sub_bytes(s_prima_matrix: pd.DataFrame):
    inv_s_box = inv_s_box_generate()
    s_array_np = []
    s_prima_sequence = matrix_to_sequence(s_prima_matrix)
    s_array = re.findall('..', s_prima_sequence)
    for byte in s_array:
        left_bits = byte[0]
        right_bits = byte[1]
        converted_value = inv_s_box.loc[left_bits, right_bits]
        s_array_np.append(converted_value)
    s_array_np = np.array(s_array_np)
    s_matrix_np = np.reshape(s_array_np, (4, 4))
    s_prima_matrix = pd.DataFrame(s_matrix_np)
    return s_prima_matrix


def circular_rotate(input: [], steps: int, left: bool):
    queries = [0, 1, 2, 3]
    max_length = len(input)
    output = []
    for q in queries:
        if left:
            output.append(input[(max_length + steps + q) % max_length])  # Left change + step
        else:
            output.append(input[(max_length - steps + q) % max_length])  # Right change - step
    return output


def shift_rows(s_matrix: pd.DataFrame):
    for i in range(0, s_matrix.index.size):
        row_array = s_matrix.iloc[i].values
        s_matrix.iloc[i] = circular_rotate(input=row_array, steps=i, left=True)
    return s_matrix


def inv_shift_rows(s_matrix: pd.DataFrame):
    for i in range(0, s_matrix.index.size):
        row_array = s_matrix.iloc[i].values
        s_matrix.iloc[i] = circular_rotate(input=row_array, steps=i, left=False)
    return s_matrix


def mix_columns(s_matrix: pd.DataFrame, m_x: int):
    mix_matrix = [[2, 3, 1, 1],
                  [1, 2, 3, 1],
                  [1, 1, 2, 3],
                  [3, 1, 1, 2]]
    mix_matrix = np.array(mix_matrix).astype('int')
    s_matrix_dec = []
    for i in range(s_matrix.index.size):
        row_dec = s_matrix.iloc[i].apply(int, base=16)
        s_matrix_dec = np.append(s_matrix_dec, row_dec, axis=0)
    s_matrix_dec = np.reshape(s_matrix_dec, (4, 4)).astype('int')

    s_prima_col = np.zeros([4, 1])
    for i in range(len(s_matrix_dec)):
        s_prima_0 = 0
        s_prima_1 = 0
        s_prima_2 = 0
        s_prima_3 = 0
        for j in range(len(s_matrix_dec)):
            s_prima_0 = addition(s_prima_0, multiplication(mix_matrix[0, j], s_matrix_dec[j][i], m_x))
            s_prima_1 = addition(s_prima_1, multiplication(mix_matrix[1, j], s_matrix_dec[j][i], m_x))
            s_prima_2 = addition(s_prima_2, multiplication(mix_matrix[2, j], s_matrix_dec[j][i], m_x))
            s_prima_3 = addition(s_prima_3, multiplication(mix_matrix[3, j], s_matrix_dec[j][i], m_x))
        s_prima_col = np.append(s_prima_col, [[format(s_prima_0, '#04x')[2:].upper()],
                                              [format(s_prima_1, '#04x')[2:].upper()],
                                              [format(s_prima_2, '#04x')[2:].upper()],
                                              [format(s_prima_3, '#04x')[2:].upper()]], axis=1)

    s_prima_matrix = s_prima_col[:, np.arange(1, len(s_matrix_dec) + 1, 1, int)]
    s_prima_matrix = pd.DataFrame(s_prima_matrix)
    return s_prima_matrix


def inv_mix_columns(s_prima_matrix: pd.DataFrame, m_x: int):
    inv_mix_matrix = [[14, 11, 13, 9],
                      [9, 14, 11, 13],
                      [13, 9, 14, 11],
                      [11, 13, 9, 14]]
    inv_mix_matrix = np.array(inv_mix_matrix).astype('int')
    s_matrix_dec = []
    for i in range(s_prima_matrix.index.size):
        row_dec = s_prima_matrix.iloc[i].apply(int, base=16)
        s_matrix_dec = np.append(s_matrix_dec, row_dec, axis=0)
    s_matrix_dec = np.reshape(s_matrix_dec, (4, 4)).astype('int')

    s_col = np.zeros([4, 1])
    for i in range(len(s_matrix_dec)):
        s_0 = 0
        s_1 = 0
        s_2 = 0
        s_3 = 0
        for j in range(len(s_matrix_dec)):
            s_0 = addition(s_0, multiplication(inv_mix_matrix[0, j], s_matrix_dec[j][i], m_x))
            s_1 = addition(s_1, multiplication(inv_mix_matrix[1, j], s_matrix_dec[j][i], m_x))
            s_2 = addition(s_2, multiplication(inv_mix_matrix[2, j], s_matrix_dec[j][i], m_x))
            s_3 = addition(s_3, multiplication(inv_mix_matrix[3, j], s_matrix_dec[j][i], m_x))
        s_col = np.append(s_col, [[format(s_0, '#04x')[2:].upper()],
                                  [format(s_1, '#04x')[2:].upper()],
                                  [format(s_2, '#04x')[2:].upper()],
                                  [format(s_3, '#04x')[2:].upper()]], axis=1)

    s_matrix = s_col[:, np.arange(1, len(s_matrix_dec) + 1, 1, int)]
    s_matrix = pd.DataFrame(s_matrix)
    return s_matrix


def add_round_key(s_matrix: pd.DataFrame, key_matrix: []):
    s_matrix_np = np.array(s_matrix)
    key_matrix_np = np.array(key_matrix)

    s_matrix_np = np.reshape(s_matrix_np, (1, 16))
    key_matrix_np = np.reshape(key_matrix_np, (1, 16))

    s_matrix_np_dec = np.zeros([1, 16])
    key_matrix_np_dec = np.zeros([1, 16])
    for i in range(len(s_matrix_np[0])):
        row_state_dec = int(s_matrix_np[0, i], 16)
        row_key_dec = int(key_matrix_np[0, i], 16)
        s_matrix_np_dec[0, i] = row_state_dec
        key_matrix_np_dec[0, i] = row_key_dec

    round_array = []
    for i in range(len(s_matrix_np_dec[0])):
        round_array = np.append(round_array, format(addition(int(s_matrix_np_dec[0, i]),
                                                             int(key_matrix_np_dec[0, i])), '#04x')[2:].upper())

    round_matrix = np.reshape(round_array, (4, 4))
    round_matrix = pd.DataFrame(round_matrix)
    return round_matrix
```

```python
# AES IMPLEMENTATION
def rc_generate(rounds: int, m_x: int):
    rc = []
    rc = np.append(rc, 0)
    rc = np.append(rc, 1)

    for i in range(2, rounds):
        rc = np.append(rc, int(multiplication(int(rc[i - 1]), 2, m_x)))

    return rc


def g_function(w: [], round: int, m_x: int):
    # RotWord
    x = circular_rotate(input=w, steps=1, left=True)

    # SubWord
    s_box = s_box_generate()
    y = []
    for byte in x:
        left_bits = byte[0]
        right_bits = byte[1]
        converted_value = s_box.loc[left_bits, right_bits]
        y = np.append(y, converted_value)

    # Rcon
    rc = rc_generate(rounds=11, m_x=m_x)
    rcon = np.append(rc[round], [0, 0, 0])
    z = []
    for i in range(len(y)):
        y_dec = int(y[i], 16)
        rcon_dec = int(rcon[i])
        z_dec = y_dec ^ rcon_dec
        z_hex = format(z_dec, '#04x')[2:].upper()
        z = np.append(z, z_hex)

    return z


def key_expansion_round_0(key_matrix: pd.DataFrame, m_x: int):
    w_array = key_matrix.values
    w_3 = key_matrix.iloc[3, :].values
    g = g_function(w=w_3, round=1, m_x=m_x)
    # w_array = np.transpose(w_array)
    return w_array, g


def key_expansion(w_array_prev: [], g: [], round: int, m_x: int):
    for i in range(4, 8):
        new = []
        if (i - 4) % 4 == 0:
            for j in range(0, 4):
                col_pev = w_array_prev[i - 4]
                xor = int(col_pev[j], 16) ^ int(g[j], 16)
                new = np.append(new, format(xor, '#04x')[2:].upper())
        else:
            for j in range(0, 4):
                col_pev = w_array_prev[i - 1]
                col_pev_2 = w_array_prev[i - 4]
                xor = int(col_pev[j], 16) ^ int(col_pev_2[j], 16)
                new = np.append(new, format(xor, '#04x')[2:].upper())
        w_array_prev = np.append(w_array_prev, new)
        w_array_prev = np.reshape(w_array_prev, (int(len(w_array_prev) / 4), 4))

    if round > 10:
        g_new = ['0', '0', '0', '0']
    else:
        g_new = g_function(w=w_array_prev[7, :], round=round, m_x=m_x)
    # w_array_final = np.transpose(w_array_prev[4:8, :])
    return w_array_prev[4:8, :], g_new


def get_all_key_expansion(key_matrix: pd.DataFrame, m_x: int):
    w_array_prev, g_prev = key_expansion_round_0(key_matrix=key_matrix, m_x=m_x)
    w_array_all = w_array_prev
    g_array_all = g_prev
    for i in range(2, 12):  # max i = 11
        w_array, g = key_expansion(w_array_prev=w_array_prev, g=g_prev, round=i, m_x=m_x)
        w_array_all = np.append(w_array_all, w_array)
        g_array_all = np.append(g_array_all, g)
        w_array_prev = w_array
        g_prev = g
    w_array_all = np.reshape(w_array_all, (int(len(w_array_all) / 4), 4))
    g_array_all = np.reshape(g_array_all, (int(len(g_array_all) / 4), 4))
    return w_array_all


def aes_encryption(plain_text_seq: str, key_seq: str, m_x: int):
    key_matrix = sequence_to_matrix(key_seq)
    w_matrix = get_all_key_expansion(key_matrix, m_x)

    plain_text_matrix = sequence_to_matrix(plain_text_seq)
    plain_text_matrix = np.transpose(plain_text_matrix)
    plain_text_matrix = pd.DataFrame(plain_text_matrix)

    # First state
    print_current_state(plain_text_matrix, round=0)
    print('w: ')
    print(np.transpose(w_matrix[0:4, :]))

    add = add_round_key(s_matrix=plain_text_matrix, key_matrix=np.transpose(w_matrix[0:4, :]))
    for i in range(1, 11):
        print_current_state(add, round=i)
        sub = sub_bytes(add)
        shift = shift_rows(sub)
        mix = mix_columns(shift, m_x)

        print('w: ')
        print(np.transpose(w_matrix[4*i:4*i + 4, :]))

        if i == 10:
            continue
        add = add_round_key(s_matrix=mix, key_matrix=np.transpose(w_matrix[4*i:4*i + 4, :]))

    # Final round
    sub = sub_bytes(add)
    shift = shift_rows(sub)
    add = add_round_key(s_matrix=shift, key_matrix=np.transpose(w_matrix[40:44, :]))
    print_current_state(add, round=11)

    cipher = matrix_to_sequence(add)

    return cipher


def aes_decryption(cipher_text_seq: str, key_seq: str, m_x: int):
    key_matrix = sequence_to_matrix(key_seq)
    w_matrix = get_all_key_expansion(key_matrix, m_x)

    cipher_text_matrix = sequence_to_matrix(cipher_text_seq)
    # cipher_text_matrix = np.transpose(cipher_text_matrix)
    cipher_text_matrix = pd.DataFrame(cipher_text_matrix)

    # First state
    round = 0
    print_current_state(cipher_text_matrix, round=round)
    add = add_round_key(s_matrix=cipher_text_matrix, key_matrix=np.transpose(w_matrix[40:44, :]))
    for i in range(1, 10):

        shift = inv_shift_rows(add)
        sub = inv_sub_bytes(shift)
        print_current_state(sub, round=i)
        print('w: ')
        print(np.transpose(w_matrix[(44 - 4 * (i+1)):(44 - 4 * (i+1)) + 4, :]))

        add = add_round_key(s_matrix=sub,
                            key_matrix=np.transpose(w_matrix[(44 - 4 * (i + 1)):(44 - 4 * (i + 1)) + 4, :]))
        mix = inv_mix_columns(add, m_x)
        add = mix

    print('End of For')
    # Final round
    # print_current_state(add, round=i)
    shift = inv_shift_rows(add)
    sub = inv_sub_bytes(shift)
    add = add_round_key(s_matrix=sub, key_matrix=np.transpose(w_matrix[0:4, :]))
    print_current_state(add, round=11)
    print('w: ')
    print(np.transpose(w_matrix[0:4, :]))

    plain = matrix_to_sequence(np.transpose(add))

    return plain


def avalanche_effect_plain_text(plain_seq_1: str, plain_seq_2: str, key_seq: str, m_x: int):
    key_matrix = sequence_to_matrix(key_seq)
    w_matrix = get_all_key_expansion(key_matrix, m_x)

    count = sum(1 for a, b in zip(plain_seq_1, plain_seq_2) if a != b)

    # First state
    print([[0, plain_seq_1, plain_seq_2], count])

    plain_matrix_1 = sequence_to_matrix(plain_seq_1)
    plain_matrix_1 = np.transpose(plain_matrix_1)
    plain_matrix_1 = pd.DataFrame(plain_matrix_1)

    plain_matrix_2 = sequence_to_matrix(plain_seq_2)
    plain_matrix_2 = np.transpose(plain_matrix_2)
    plain_matrix_2 = pd.DataFrame(plain_matrix_2)

    add_1 = add_round_key(s_matrix=plain_matrix_1, key_matrix=np.transpose(w_matrix[0:4, :]))
    add_2 = add_round_key(s_matrix=plain_matrix_2, key_matrix=np.transpose(w_matrix[0:4, :]))
    for i in range(1, 11):
        add_1_seq = matrix_to_sequence(np.transpose(add_1))
        add_2_seq = matrix_to_sequence(np.transpose(add_2))
        count = sum(1 for a, b in zip(add_1_seq, add_2_seq) if a != b)
        print([[i-1, add_1_seq, add_2_seq], count])

        sub_1 = sub_bytes(add_1)
        shift_1 = shift_rows(sub_1)
        mix_1 = mix_columns(shift_1, m_x)

        sub_2 = sub_bytes(add_2)
        shift_2 = shift_rows(sub_2)
        mix_2 = mix_columns(shift_2, m_x)

        if i == 10:
            continue
        add_1 = add_round_key(s_matrix=mix_1, key_matrix=np.transpose(w_matrix[4*i:4*i + 4, :]))
        add_2 = add_round_key(s_matrix=mix_2, key_matrix=np.transpose(w_matrix[4 * i:4 * i + 4, :]))

    # Final round
    sub_1 = sub_bytes(add_1)
    shift_1 = shift_rows(sub_1)
    add_1 = add_round_key(s_matrix=shift_1, key_matrix=np.transpose(w_matrix[40:44, :]))

    sub_2 = sub_bytes(add_2)
    shift_2 = shift_rows(sub_2)
    add_2 = add_round_key(s_matrix=shift_2, key_matrix=np.transpose(w_matrix[40:44, :]))

    add_1_seq = matrix_to_sequence(np.transpose(add_1))
    add_2_seq = matrix_to_sequence(np.transpose(add_2))
    count = sum(1 for a, b in zip(add_1_seq, add_2_seq) if a != b)
    print([[10, add_1_seq, add_2_seq], count])


def avalanche_effect_key(plain_text_seq: str, key_seq_1: str, key_seq_2: str, m_x: int):
    key_matrix_1 = sequence_to_matrix(key_seq_1)
    w_matrix_1 = get_all_key_expansion(key_matrix_1, m_x)

    key_matrix_2 = sequence_to_matrix(key_seq_2)
    w_matrix_2 = get_all_key_expansion(key_matrix_2, m_x)

    plain_text_matrix = sequence_to_matrix(plain_text_seq)
    plain_text_matrix = np.transpose(plain_text_matrix)
    plain_text_matrix = pd.DataFrame(plain_text_matrix)

    # First state
    count = sum(1 for a, b in zip(plain_text_seq, plain_text_seq) if a != b)

    # First state
    print([[0, plain_text_seq, plain_text_seq], count])

    add_1 = add_round_key(s_matrix=plain_text_matrix, key_matrix=np.transpose(w_matrix_1[0:4, :]))
    add_2 = add_round_key(s_matrix=plain_text_matrix, key_matrix=np.transpose(w_matrix_2[0:4, :]))
    for i in range(1, 11):
        add_1_seq = matrix_to_sequence(np.transpose(add_1))
        add_2_seq = matrix_to_sequence(np.transpose(add_2))
        count = sum(1 for a, b in zip(add_1_seq, add_2_seq) if a != b)
        print([[i-1, add_1_seq, add_2_seq], count])

        sub_1 = sub_bytes(add_1)
        shift_1 = shift_rows(sub_1)
        mix_1 = mix_columns(shift_1, m_x)

        sub_2 = sub_bytes(add_2)
        shift_2 = shift_rows(sub_2)
        mix_2 = mix_columns(shift_2, m_x)

        if i == 10:
            continue
        add_1 = add_round_key(s_matrix=mix_1, key_matrix=np.transpose(w_matrix_1[4 * i:4 * i + 4, :]))
        add_2 = add_round_key(s_matrix=mix_2, key_matrix=np.transpose(w_matrix_2[4 * i:4 * i + 4, :]))

        # Final round
    sub_1 = sub_bytes(add_1)
    shift_1 = shift_rows(sub_1)
    add_1 = add_round_key(s_matrix=shift_1, key_matrix=np.transpose(w_matrix_1[40:44, :]))

    sub_2 = sub_bytes(add_2)
    shift_2 = shift_rows(sub_2)
    add_2 = add_round_key(s_matrix=shift_2, key_matrix=np.transpose(w_matrix_2[40:44, :]))

    add_1_seq = matrix_to_sequence(np.transpose(add_1))
    add_2_seq = matrix_to_sequence(np.transpose(add_2))
    count = sum(1 for a, b in zip(add_1_seq, add_2_seq) if a != b)
    print([[10, add_1_seq, add_2_seq], count])
```
