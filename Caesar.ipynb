```python
def integer_convert(input: str):
    output = 0
    try:
        output = int(input)
    except:
        print('Validation: Please enter only numbers')
    return output

def text_processing(input: str):
    ascii_value = 0
    caesar_mapping = 65 # In Caesar is 0 but in ASCII is 65
    output = ""
    for character in input:
        ascii_value = ord(character)        
        if ascii_value < (0 + caesar_mapping) or ascii_value > (25 + caesar_mapping):
            print('Validation: Please enter only capital letters')
            print('Validation: Error in the letter: ', character)
            raise Exception("Sorry, Please enter only capital letters")
        output = output + character
    return output

def caesar_encryption(plain_text: str, key: int):
    ciphertext = ""
    caesar_mapping = 65 # In Caesar is 0 but in ASCII is 65
    str_encrypted = ""
    for character in plain_text:
        caesar_value = ord(character)
        ascii_value = caesar_value - caesar_mapping
        ascii_encrypted = (ascii_value + key) % 26
        caesar_encrypted = ascii_encrypted + caesar_mapping        
        str_encrypted = chr(caesar_encrypted)
        ciphertext = ciphertext + str_encrypted
    return ciphertext

def caesar_decryption(ciphertext: str, key: int):
    plain_text = ""
    caesar_mapping = 65 # In Caesar is 0 but in ASCII is 65
    str_decrypted = ""
    for character in ciphertext:
        caesar_value = ord(character)
        ascii_value = caesar_value - caesar_mapping
        ascii_decrypted = (ascii_value - key) % 26
        caesar_decrypted = ascii_decrypted + caesar_mapping
        str_decrypted = chr(caesar_decrypted)
        plain_text = plain_text + str_decrypted
    return plain_text

def frequency_calculation(ciphertext: str):
    frequency = dict()
    for character in ciphertext:
        if character in frequency.keys():
            frequency[character] = frequency[character] + 1
        else:
            frequency[character] =  1
    return frequency
```

```python
#Imports
from matplotlib import pyplot as plt
# Principal function
# Options
EXIT = 9
#OPTIONS = [EXIT, WRITE_TEXT]
key = 3

option = 0
while(option != EXIT):
    option = input('Select one option: \n 1. Encryption \n 2. Decryption \n 3. Frequency \n 9. Exit \n')
    option = integer_convert(option)
    method = option
    if option == 0 or option == 9:
        continue
    
    option = input('Select one option: \n 1. Insert text by console \n 9. Exit \n')
    option = integer_convert(option)
    if option == 0 or option == 9:
        continue
    elif option == 1:        
        if method == 1: # Encryption
            plain_text = ""
            write_status = True
            option_write = 0
            
            # Validation processing
            while(option_write != EXIT): # Write text cycle
                plain_text = input('Write the plain text: \n')
                try:
                    plain_text = text_processing(plain_text)
                    break
                except:
                    option_write = input('Select one option: \n 1. Try again \n 9. Exit \n')
                    option_write = integer_convert(option_write)
                    if option_write == 0 | option_write == 9:
                        continue
            
            if option_write == 9:
                continue
            
            key = input('Insert key value to Caesar: \n')
            key = integer_convert(key)
            
            # TODO: Validate Key max value 25
            
            ciphertext = caesar_encryption(plain_text, key)
            print('-----------------------------------')
            print('The ciphertext is: \n', ciphertext)
            print('-----------------------------------')
            
        elif method == 2: # Decryption
            ciphertext = ""
            write_status = True
            option_write = 0
            
            # Validation processing
            while(option_write != EXIT): # Write text cycle
                ciphertext = input('Write the ciphertext: \n')
                try:
                    ciphertext = text_processing(ciphertext)
                    break
                except:
                    option_write = input('Select one option: \n 1. Try again \n 9. Exit \n')
                    option_write = integer_convert(option_write)
                    if option_write == 0 | option_write == 9:
                        continue
            
            if option_write == 9:
                continue
            
            key = input('Insert key value to Caesar: \n')
            key = integer_convert(key)
            
            # TODO: Validate Key max value 25
            
            plain_text = caesar_decryption(ciphertext, key)
            print('-----------------------------------')
            print('The plain text is: \n', plain_text)
            print('-----------------------------------')
            
        elif method == 3: # Frequency
            ciphertext = ""
            write_status = True
            option_write = 0
            
            # Validation processing
            while(option_write != EXIT): # Write text cycle
                ciphertext = input('Write the ciphertext: \n')
                try:
                    ciphertext = text_processing(ciphertext)
                    break
                except:
                    option_write = input('Select one option: \n 1. Try again \n 9. Exit \n')
                    option_write = integer_convert(option_write)
                    if option_write == 0 | option_write == 9:
                        continue
            
            if option_write == 9:
                continue
                
            frequency = frequency_calculation(ciphertext)
            frequency = frequency.items()
            x, y = zip(*frequency) 
            
            plt.bar(x, y)
            plt.xlabel('Letters')
            plt.ylabel('Frequency')
            plt.show()
    
    
    option = input('Select one option: \n 1. Continue \n 9. Exit \n')
    option = integer_convert(option)    
    
print('Script finished')
    
```
